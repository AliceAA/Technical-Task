```python
# import libraries

from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# generate initial dataset that contains "n_points_initial" number of points
# since at task density of session time is shown, 
# let's use a gamma distribution for generating random values

n_points_initial = 10000
data = np.array(stats.gamma.rvs(2.5, size=n_points_initial))
```

```python
# calculate a standart diviation and plot a distribution of an initial values

sns.set_style('darkgrid')
plt.xlim(0, 15)

print(np.std(data))
sns.distplot(data)
```

```python
# define a function that generate new values: 
# for n_times it choose n_samples number of points from data and calculate their mean,
# new values are exactly this means.


def sample_distribution (data, n_samples, n_times):
    mean_values = np.empty(n_times)
    for i in range(n_times):
        mean_values[i] = np.random.choice(data, n_samples).mean()
    return mean_values
```

```python
# calculating values that produced as means of different number of points (1, 10, 100, 1000)
# selected from the initial distribution fixed same large number of times (10000)

sample_1 = sample_distribution(data, 1, 10000)
sample_10 = sample_distribution(data, 10, 10000)
sample_100 = sample_distribution(data, 100, 10000)
sample_1000 = sample_distribution(data, 1000, 10000)
```

```python
# plot distributions of generated values
# we can notice that choosing 1 element we receive destribution close to initial one
# by choosing more points for averaging we received distributin close to normal with higher peak, 
# meaning lower standart diviation


f, axes = plt.subplots(2, 2, figsize=(15, 10), sharex = True)

sns.distplot( sample_1 , color="skyblue", ax=axes[0, 0])
sns.distplot( sample_10 , color="olive", ax=axes[0, 1])
sns.distplot( sample_100 , color="gold", ax=axes[1, 0])
sns.distplot( sample_1000 , color="teal", ax=axes[1, 1])
```

```python
# standart diviation for generated values;
# std goes down as numbers of values for averaging go up

print("Statdart diviation for means of 1 value: ", np.std(sample_1))
print("Statdart diviation for means of 10 value: ", np.std(sample_10))
print("Statdart diviation for means of 100 value: ", np.std(sample_100))
print("Statdart diviation for means of 1000 value: ", np.std(sample_1000))
```
